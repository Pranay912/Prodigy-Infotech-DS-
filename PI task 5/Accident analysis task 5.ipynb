```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('Accidents.csv')
```

```python
plt.rcParams.update({'font.size': 14})
# To set the font size of the whole notebook
```

```python
fig, axes = plt.subplots(nrows=6, ncols=2, figsize=(20, 25))
axes = axes.flatten()

for i, col in enumerate(data.columns[:-6]):
    sns.countplot(data=data, y=col, ax=axes[i], palette='tab20',)
    for p in axes[i].patches:
        width = p.get_width()
        axes[i].text(width + 0.1, p.get_y() + p.get_height() / 2., int(width), ha="left", va="center")

for i in range(len(data.columns[:-6]), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```

```python
# 18-50 age group has more accidents
# most of the drivers are male.
# Suprisingly people who had education level of junior high school caused more accidents. Education is important.
# Light condition are not so significant.
# Experience drivers too caused accidents.
# Asphalt roads saw most of the accidents.
```

```python
ax=sns.countplot(x=data['Accident_severity'])
for p in ax.patches:
    height = p.get_height()
    # Add text label above each bar
    ax.text(p.get_x() + p.get_width() / 2., height + 0.1,  # Adjust vertical position of the label
            f'{int(height)}',  # Convert height to integer
            ha='center', va='bottom')

plt.show()
```
